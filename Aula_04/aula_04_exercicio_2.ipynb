```python
from google.colab import userdata
from groq import Groq
import gradio as gr
```

```python
GROQ_API = userdata.get('appLando')
```

```python
client = Groq(api_key=GROQ_API)
```

```python
def assistente_agricultura(user_prompt: str, temperatura: float) -> str:
    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": "Você é um assistente de agricultor"},
            {"role": "user", "content": user_prompt}
        ],
        temperature=temperatura,  # temperatura variável
        max_tokens=1024
    )
    return completion.choices[0].message.content

with gr.Blocks(title="Exercício 02 – Teste de Temperatura") as demo:

    gr.Markdown("## 🌡️ Teste de Temperatura no Modelo LLM")

    entrada = gr.Textbox(label="Pergunta", lines=8)
    temperatura = gr.Slider(0, 1, value=0.5, label="Temperatura")
    saida = gr.Textbox(label="Resposta", lines=10)

    gr.Button("Enviar").click(
        fn=assistente_agricultura,
        inputs=[entrada, temperatura],
        outputs=saida
    )

demo.launch()
```
