```python
from google.colab import userdata
from groq import Groq
import gradio as gr
```

```python
GROQ_API = userdata.get('appLando')
```

```python
client = Groq(api_key=GROQ_API)
```

```python
def assistente_agricultura(user_prompt: str) -> str:
    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # modelo alterado
        messages=[
            {"role": "system", "content": "Você é um assistente de agricultor"},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0,
        max_tokens=1024
    )
    return completion.choices[0].message.content

demo = gr.Interface(
    fn=assistente_agricultura,
    inputs=gr.Textbox(lines=8, label="Pergunta"),
    outputs=gr.Textbox(lines=10, label="Resposta"),
    title="Exercício 01 – Teste de Modelo"
)

demo.launch()
```
