```python
from google.colab import userdata
from groq import Groq
import gradio as gr
```

```python
# Carrega a chave da API salva no Colab
GROQ_API = userdata.get('appLando')
```

```python
//print(GROQ_API is not None)
```

```python
# Cria o cliente Groq
client = Groq(api_key=GROQ_API)
```

```python
# Função do chatbot
def assistente_agricultura(user_prompt: str) -> str:
    try:
        if user_prompt.strip() == "15":
            return "Encerrando o Chatbot"

        completion = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {"role": "system", "content": "Você é um assistente de agricultor"},
                {"role": "user", "content": user_prompt}
            ],
            temperature=0.5,
            max_tokens=1024
        )

        return completion.choices[0].message.content

    except Exception as e:
        return f"Erro: {e}"


# ============================================================
# INTERFACE EM PORTUGUÊS (GRADIO BLOCKS)
# ============================================================

with gr.Blocks(title="Assistente de Agricultura") as demo:

    gr.Markdown("## 🌱 Assistente de Agricultura")
    gr.Markdown("Digite sua pergunta sobre agricultura e receba respostas da IA especialista.")

    entrada = gr.Textbox(
        label="Pergunta",
        placeholder="Digite sua pergunta aqui...",
        lines=8
    )

    saida = gr.Textbox(
        label="Resposta",
        lines=10
    )

    with gr.Row():
        botao_enviar = gr.Button("Enviar")
        botao_limpar = gr.Button("Limpar")

    # Ações dos botões
    botao_enviar.click(fn=assistente_agricultura, inputs=entrada, outputs=saida)
    botao_limpar.click(fn=lambda: ("", ""), inputs=None, outputs=[entrada, saida])

# Executa a interface
demo.launch(debug=True, show_error=True)
```
