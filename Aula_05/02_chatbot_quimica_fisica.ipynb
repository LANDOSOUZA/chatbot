```python
# Recupera a chave da API do Groq salva nos Secrets do Colab
GROQ_API_KEY = userdata.get("appLando")

# Verificação básica
if GROQ_API_KEY:
    print("✅ API Key do Groq carregada com sucesso")
else:
    print("❌ API Key do Groq NÃO encontrada")
```

```python

# Cria o cliente Groq
client = Groq(api_key=GROQ_API_KEY)

print("✅ Cliente Groq pronto para uso")
```

```python
def assistente_quimica_fisica(pergunta: str) -> str:
    try:
        if pergunta.strip() == "15":
            return "Encerrando o chatbot."

        resposta = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "system",
                    "content": "Você é um assistente especialista em química e física. "
                               "Responda de forma clara, objetiva e didática."
                },
                {
                    "role": "user",
                    "content": pergunta
                }
            ],
            temperature=0.3,
            max_tokens=512,
            top_p=1,
            stream=False
        )

        return resposta.choices[0].message.content

    except Exception as e:
        return f"Erro no chatbot: {e}"
```

```python
print(assistente_quimica_fisica("O que é a Primeira Lei de Ohm?"))
```

```python
iface = gr.Interface(
    fn=assistente_quimica_fisica,
    inputs=gr.Textbox(
        lines=5,
        placeholder="Digite sua pergunta de Química ou Física (15 para sair)"
    ),
    outputs=gr.Textbox(
        lines=10,
        label="Resposta do ChatBot"
    ),
    title="ChatBot Especialista em Química e Física",
    description="Digite uma pergunta sobre conceitos de Química ou Física e receba respostas da IA."
)

iface.launch(debug=True)
```
